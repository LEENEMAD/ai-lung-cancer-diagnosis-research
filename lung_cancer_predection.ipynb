# importing the needed libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

# importing the dataset and exploring it

```python
df = pd.read_csv('/content/survey lung cancer (2).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# preprocessing

encoding

```python
# I encoded all the columns to convert the categorical columns to numerical columns and change the numerical columns values from 1,2 to 0,1 to make the dataset more organized
for col in df.columns:
    df[col] = LabelEncoder().fit_transform(df[col])
```

normlization

```python
df['AGE'] = MinMaxScaler().fit_transform(df[['AGE']])
```

dropping duplicates

```python
df.drop_duplicates(inplace=True)
```

dataframe after normlization and encoding

```python
df.head()
```

corelation between features

```python
plt.figure(figsize=(18, 18))
sns.heatmap(df.corr(), annot=True, vmin = -1,vmax=1)
plt.show()
```

setting x and y

```python
X = df.drop(['LUNG_CANCER'],axis=1)
y = df['LUNG_CANCER']
```

# KNN classification model

```python
X_train, X_test, y_train, y_test_knn = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

kNN_class = KNeighborsClassifier(n_neighbors=3)

kNN_class.fit(X_train, y_train)

knn_y_predict = kNN_class.predict(X_test)
```

evaluation

```python
knn_accuracy = accuracy_score(y_test_knn, knn_y_predict)
knn_precision = precision_score(y_test_knn, knn_y_predict)
knn_recall = recall_score(y_test_knn, knn_y_predict)
knn_f1 = f1_score(y_test_knn, knn_y_predict)
print(f'The Accuracy of KNN: {knn_accuracy}')
print(f'The precision of KNN: {knn_precision}')
print(f'The recall of KNN: {knn_recall}')
print(f'The F1 score KNN: {knn_f1}')
```

confusion_matrix

```python
confusion_matrix(y_test_knn, knn_y_predict)
```

# Decision tree classifier

```python
X_train, X_test, y_train, y_test_dt = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

dt_class = DecisionTreeClassifier()

dt_class.fit(X_train, y_train)

dt_y_predict = dt_class.predict(X_test)
```

evaluation

```python
dt_accuracy = accuracy_score(y_test_dt, dt_y_predict)
dt_precision = precision_score(y_test_dt, dt_y_predict)
dt_recall = recall_score(y_test_dt, dt_y_predict)
dt_f1 = f1_score(y_test_dt, dt_y_predict)

print(f'The Accuracy of Decision Tree: {dt_accuracy}')
print(f'The precision of Decision Tree: {dt_precision}')
print(f'The recall of Decision Tree: {dt_recall}')
print(f'The F1 score Decision Tree: {dt_f1}')
```

confusion_matrix

```python
confusion_matrix(y_test_dt, dt_y_predict)
```

# Random forest classifier

```python
X_train, X_test_rf, y_train, y_test_rf = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

rf_y_predict = rf.predict(X_test)
```

evaluation

```python
rf_accuracy = accuracy_score(y_test_rf, rf_y_predict)
rf_precision = precision_score(y_test_rf, rf_y_predict)
rf_recall = recall_score(y_test_rf, rf_y_predict)
rf_f1 = f1_score(y_test_rf, rf_y_predict)

print(f'The Accuracy of Decision Tree: {rf_accuracy}')
print(f'The precision of Decision Tree: {rf_precision}')
print(f'The recall of Decision Tree: {rf_recall}')
print(f'The F1 score Decision Tree: {rf_f1}')
```

confusion_matrix

```python
confusion_matrix(y_test_rf, rf_y_predict)
```
